## 2, Import Librarys dan Konfigurasi awal

```python
from google_play_scraper import reviews_all, Sort 
# jadi disini fungsi dari reviews_all adalah untuk menarik reviews dari aplikasi yang kita mau 
# untuk Sort disini fungsinya adalah konstanta yang berfungsi mengurutkan reviews
import pandas as pd
# menyimpan data dalam bentuk data frame 
import time
# untuk mengukur lama scrapping 

APP_ID = "com.jago.digitalBanking"
OUTPUT_CSV = "../csv/jago_reviews_raw.csv"
```

```python
def data_scrap (app_id: str, langs: list[str], country: str = "id") -> pd.DataFrame:
    #ini buat namain fungsinya + declare outpurnya nanti dalam bentuk string dan mengambil di indonesia.
    all_dfs=[]
    for lang in langs:
        print(f"Bahasa dari data = {lang}, diambil dari negara {country}")
        start = time.time() # buat ngukur waktu scrap
        result = reviews_all( #manggil fungsi reviews all yang udah kita import tadi dan kita harus isi parameternya
            app_id,
            lang=lang,
            country=country,
            sort= Sort.NEWEST,
            sleep_milliseconds = 50,
        )
        elapsed = time.time() - start #ini ngitung waktu selisih dari start
        print(f"waktu scrap{elapsed:.2f}")
        print(f"Data yang terambil: {len(result)}")


        df_lang = pd.DataFrame(result)
        df_lang["lang"] = lang #nambah kolom lang buat tagging bahasa
        all_dfs.append(df_lang) #masukin ke array all_dfs


    df_all = pd.concat(all_dfs, ignore_index=True) #gabungin df di all_dfs dengan index baru
    print(f"Total gabungan semua bahasa: {len(df_all)} review")
    return df_all
```

```python
df_jago = data_scrap (
    app_id = APP_ID,
    langs = ["id","en"],
    country = "id"
    )
df_jago.head() # lihat kolomnya udah bener apa ngga
df_jago["lang"].value_counts() # ini buat ngelihat jumlah komen per bahasa dikolom lang
df_jago.to_csv(OUTPUT_CSV, index=False)
```
